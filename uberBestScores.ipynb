```python
df["pickup_datetime"] = pd.to_datetime(df["pickup_datetime"])
df["hours"] = df["pickup_datetime"].dt.hour
df["day_of_the_week"] = df["pickup_datetime"].dt.day_of_week
```

```python
def calculateDistance(lat1, lon1, lat2, lon2):
    R = 6371
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat/2) ** 2 + np.cos(lat1/2) * np.cos(lat2/2) * np.sin(dlon/2) ** 2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))

    distance = R * c
    return distance
```

```python
df["Distance"] = calculateDistance(df["pickup_latitude"], df["pickup_longitude"], df["dropoff_latitude"], df["dropoff_longitude"])
```

```python
for col in df.columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lb = Q1 - 1.5 * (IQR)
    ub = Q3 + 1.5 * (IQR)
    df = df[(df[col] >= lb) & (df[col] <= ub)]
```

```python
plt.figure(figsize=(15, 10))
sns.boxplot(df)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, root_mean_squared_error, r2_score, mean_squared_error
```

```python
X = df[['fare_amount', 'passenger_count', 'hours', 'day_of_the_week', 'Distance']]
y = df['fare_amount']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
r2score = r2_score(y_test, y_pred)
print("R2 Score: ", r2score)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print("RMSE: ", rmse)
```

```python
from sklearn.ensemble import RandomForestRegressor
model1 = RandomForestRegressor()
model1.fit(X_train, y_train)
```

```python
y_pred = model1.predict(X_test)
```

```python
r2score = r2_score(y_test, y_pred)
print("R2 Score: ", r2score)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print("RMSE: ", rmse)
```
